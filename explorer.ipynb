```python
root_regex = '\(Root\s\(span\s(\d*)\s(\d*))\) '


root_regex = '.*(Root).*\(span\s(\d*)\s(\d*).*'
nucleus_span_regex = '.*(Nucleus|Satellite).*\((span)\s(\d*)\s(\d*).*\(rel2par(.*?)\)'
nucleus_leaf_regex = '.*(Nucleus|Satellite).*\((leaf)\s(\d*).*\(rel2par(.*?)\).*\(text\s(.*?)\s\)'
satellite_span_regex = '.*(Satellite).*\(span\s(\d*)\s(\d*).*\(rel2par(.*?)\)'
satellite_leaf_regex = '.*(Satellite).*\(leaf\s(\d*).*\(rel2par(.*?)\)'

```

```python
rst_tree = []

for line in data.split('\n') : 


    if re.findall(root , line) : 
        rst_tree.append(re.findall(root , line)[0])
    elif re.findall(nucleus_span_regex , line) : 
        rst_tree.append(re.findall(nucleus_span_regex , line)[0])
    elif re.findall(nucleus_leaf_regex , line) :
        rst_tree.append(re.findall(nucleus_leaf_regex , line)[0])
    else : 
        pass
```

```python
rst = rst_tree.copy()
for r in rst : 
    print(r)
```

```python
def get_vals(node) : 

    vals = {'type' : node[0]}

    if vals['type'] == 'Root'  : 

        vals['i'] = int(node[1])
        vals['j'] = int(node[2])
        vals['rel'] = None
        vals['text'] = None

    else : 

        if node[1] == 'span' : 
            vals['i'] = int(node[2])
            vals['j'] = int(node[3])
            vals['rel'] = node[4].strip()
            vals['text'] = None

        else : 
            
            # if vals['type'] == 'Nucleus' :
            #     vals['i'] = int(node[2])
            # else : 
            #     vals['j'] = int(node[2])
            vals['i'] = int(node[2])
            vals['j'] = int(node[2])
            vals['rel'] = node[3].strip()
            vals['text'] = node[4].strip()

    return vals
```

```python
indices = []

for r in rst : 
    vals = get_vals(r)
    indices.append(vals)


indices[:10]
```

```python
relations = []

for cnt , r in enumerate(indices) : 

    curr_i = r['i']
    curr_j = r['j']

    candidates = []

    largest_i_span = 0
    largest_j_span = 0

    if r['rel'] != 'Same-Unit' and curr_i!=curr_j : 

        for r_ in indices[cnt+1 : ] :
            
            if r_['i'] == curr_i and  (r_['j']-r_['i'] >= largest_i_span) : 
                candidates.append(r_)
                largest_i_span = r_['j']-r_['i']

            if r_['j']==curr_j and (r_['j']-r_['i'] >= largest_j_span): 
                candidates.append(r_)
                largest_j_span = r_['j']-r_['i']

        r['span_len'] = curr_j - curr_i
        relations.append((r , candidates))

```

```python
for x in relations : 
    print(x)
```

```python
sorted_relations = sorted(relations , key=lambda x : x[0]['span_len'])
```

```python
x = [leaves[i+1] for i in range(len(leaves))]


for sr in sorted_relations : 

    i = sr[0]['i']-1
    j = sr[0]['j']-1

    if len(sr[1]) == 2 :
        relation = sr[1][1]['rel']
        k = sr[1][1]['i']-1
        l = sr[1][0]['j']-1
        x[l] = '{} </{}>'.format(x[l] , relation)
        x[k] = '</{}> {} '.format(relation , x[k] )
        x[i] = '<{}> {}'.format(relation , x[i])
        x[j] = '{} <//{}>'.format(x[j] , relation )

    else : 
        print(sr[1])


```

```python
stack = 0
prettified = []

combined = ' '.join(x).split(' ')

for c in combined : 

    if c[:3] == '<//' : 
        prettified.append(' '*stack + c)
        stack -= 1 

    elif c[:2] == '</' : 
        prettified.append(' '*stack + c)

    elif c[:1] == '<' : 
        stack += 1 
        prettified.append(' '*stack + c)

    else : 
        prettified.append('  '*stack + c)


for p in prettified : 
    print(p)
```
